# Compute Cost

```python
X_train.shape[0]
```

```python
def compute_cost(x, y, w, b):
    """
    x = Median Income
    y = Median House Value
    w, b = parameters of he model
    
    """
    
    m = x.shape[0]
    total_cost = 0
    
    cost = 0
    for i in range(m):
        f_wb = w*x[i] + b
        cost += (f_wb - y[i])**2
    total_cost = cost / (2*m)
    
    return total_cost
    
    
```

```python
initial_w = 0
initial_b = 0

cost = compute_cost(X_train, y_train, initial_w, initial_b)
print(type(cost))
print(f'Cost at initial w (zeros): {cost[0]:.3f}')
```

# Using Gradient Descent to find Min Cost and parameters

```python
import copy
import math
%matplotlib inline
```

```python
def compute_gradient(x, y, w, b):
    m = x.shape[0]
    
    dj_dw = 0
    dj_db = 0
    
    for i in range(m):
        f_wb = w*x[i] + b
        dj_db += f_wb - y[i]
        dj_dw += (f_wb - y[i]) * x[i]
    dj_dw /= m
    dj_db /= m
    
    return dj_dw, dj_db
```

```python
def gradient_descent(x, y, w_in, b_in, cost_function, gradient_function, alpha, iterations):
    m = len(x) #num of training examples
    
    J_history = [] # Store for printing iterations / graphing
    w_history = []
    w = copy.deepcopy(w_in)
    b = b_in
    
    for i in range(iterations):
        dj_dw, dj_db = gradient_function(x, y, w, b) # caculate gradient
        w = w - alpha * dj_dw # update parameters
        b = b - alpha * dj_db
        
        if i < 100000: # prevent resouce exhaustion
            cost = cost_function(x, y, w, b)
            J_history.append(cost[0]) 
    
        if i % math.ceil(iterations / 10) == 0:
            w_history.append(w)
            print(f'Iteration {i:4}: Cost {float(J_history[-1]):8.2f}')
    return w, b, J_history, w_history
```

```python
initial_w = 0
initial_b = 0

# some gradient descent settings
iterations = 1500
alpha = 0.03

w,b,_,_ = gradient_descent(X_train ,y_train, initial_w, initial_b, compute_cost, compute_gradient, alpha, iterations)
print("w,b found by gradient descent:", w, b)
```

```python
m = X_train.shape[0]
predicted = np.zeros(m)

for i in range(m):
    predicted[i] = w * X_train[i] + b
```

```python
# Plot the linear fit
plt.plot(X_train, predicted, c = "b")

# Create a scatter plot of the data. 
plt.scatter(X_train, y_train, marker='x', c='r') 

plt.title('Median House Value vs. Median Income')
plt.ylabel('Median House Value')
plt.xlabel('Median Income (10,000s)')
```
